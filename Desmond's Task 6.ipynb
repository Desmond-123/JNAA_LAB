```python
import numpy as np
import matplotlib as mpl
from matplotlib import pyplot as plt
import pandas as pd
```

```python
a=pd.read_excel('C:/Users/USER/Desktop/WEATHER CALCULATOR.xlsx')
```

```python
a
```

```python
fig, axes = plt.subplots(figsize=(10,6))

ax1=axes
ax1.plot(a.index,a.TEMP,color='red',lw=2, marker='*', markersize=8)
ax1.set_xlabel('TIME [UTC]', fontsize=20)
ax1.set_ylabel('TEMPERATURE [$^o$ C]', fontsize=20)

axes.legend()

ax2=axes.twinx()
ax2.plot(a.index,a.RH,color='blue',lw=2, marker='o', markersize=8)
ax2.set_xlabel('Time UTC',fontsize=20)
ax2.set_ylabel('RELATIVE HUMIDITY [%]', fontsize=20)


plt.title('Interaction of RH and Temp.', fontsize=15)
```
